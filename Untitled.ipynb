```python
import numpy as np

# Randomly add Gender and Married columns
np.random.seed(42)
df['Gender'] = np.random.choice(['Male', 'Female'], size=len(df))
df['Married'] = np.random.choice(['Yes', 'No'], size=len(df))

# Now this will work
df.groupby(['Married', 'Gender']).mean(numeric_only=True)['installment']
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Function to encode categorical columns
def encode_labels(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
    return data

# Apply encoding
df = encode_labels(df)

# Plot heatmap of correlation > 0.8
plt.figure(figsize=(12, 8))
sb.heatmap(df.corr(numeric_only=True) > 0.8, annot=True, cbar=False)
plt.title("Highly Correlated Feature Pairs (r > 0.8)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

# Split features and target
features = df.drop('not.fully.paid', axis=1)
target = df['not.fully.paid'].values

# Train-test split
X_train, X_val, Y_train, Y_val = train_test_split(
    features, target, test_size=0.2, random_state=10
)

# Handle class imbalance
ros = RandomOverSampler(sampling_strategy='minority', random_state=0)
X, Y = ros.fit_resample(X_train, Y_train)

# Shape comparison
print("Original training shape:", X_train.shape)
print("After oversampling shape:", X.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize training features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Normalize validation features using the same scaler
X_val = scaler.transform(X_val)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC

model = SVC(kernel='rbf', probability=True)  # Add probability=True if needed later for prob-based ROC curve

model.fit(X, Y)

# Evaluate using ROC AUC
print('Training AUC :', roc_auc_score(Y, model.predict(X)))
print('Validation AUC :', roc_auc_score(Y_val, model.predict(X_val)))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, confusion_matrix

# ROC AUC Scores using class predictions
training_roc_auc = roc_auc_score(Y, model.predict(X))
validation_roc_auc = roc_auc_score(Y_val, model.predict(X_val))

print('Training ROC AUC Score:', training_roc_auc)
print('Validation ROC AUC Score:', validation_roc_auc)
print()

# Confusion Matrix
cm = confusion_matrix(Y_val, model.predict(X_val))
print('Confusion Matrix:\n', cm)
```

```python
plt.figure(figsize=(6, 6))
sb.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_val, model.predict(X_val)))
```
